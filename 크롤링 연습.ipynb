#### metacharacters (메타 캐릭터)
- [] 문자들의 범위를 나타내기 위해 사용

- [abck] : a or b or c or k
- [abc.^] : a or b or c or . or ^
- [a-d] : -와 함께 사용되면 해당 문자 사이의 범위에 속하는 문자 중 하나
- [0-9] : 모든 숫자
- [a-z] : 모든 소문자
- [A-Z] : 모든 대문자
- [a-zA-Z0-9] : 모든 알파벳 문자 및 숫자
- [^0-9] : ^가 맨 앞에 사용 되는 경우 해당 문자 패턴이 아닌 것과 매칭
#### 다른 문자와 함께 사용되어 특수한 의미로 사용

- \d : 숫자를 [0-9]와 동일
- \D : 숫자가 아닌 문자 [^0-9]와 동일
- \s : 공백 문자(띄어쓰기, 탭, 엔터 등)
- \S : 공백이 아닌 문자
- \w : 알파벳대소문자, 숫자 [0-9a-zA-Z]와 동일
- \W : non alpha-numeric 문자 [^0-9a-zA-Z]와 동일
- \t, \n, \r - tab, newline, return
####  $ , ^

- ^ : 문자열의 맨 앞부터 일치하는 경우 검색

- $ : 문자열의 맨 뒤부터 일치하는 경우 검색

- ^ = 시작, $ = 끝 : 각각 문자열의 시작과 끝을 의미

```python
print('case01 - bs ')
print()

# 보통 request 하나를 쓰는데 request.post를 사용했다. request.post는 url, dict(json,) 2개의 특성을 가진다. 
# queryString 을 딕트로 선언하고 그 밑에 딕트 유형을 선언했다. 
# 이걸 못알아봤네... 

url = 'http://www.cine21.com/rank/person/content'
queryString = dict()
queryString['section']      = 'actor'
queryString['period_start'] = '2022-03'
queryString['gender']       = 'all'
queryString['page']         = '1'

# 선언한 딕트타입을 확인해봅시다. 
print(queryString)
```

```python
response = requests.post(url , queryString) 
response
# response.content 뒤에 content는 무슨의미 ?? 
# 웹크롤링시 response.text , response.content의 차이점? 
# 유니코드로 반환하는 것과 바이트로 반환하는것은 무슨차이 ? 
soup = BeautifulSoup(response.content, 'html.parser')
soup
```
