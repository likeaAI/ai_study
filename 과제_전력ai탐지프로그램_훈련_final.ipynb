```python
#1차 모델 - 실패

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from google.colab import drive

# Google Drive와 연동
drive.mount('/content/drive')

# 데이터 경로
train_path = '/content/drive/MyDrive/train.csv'

test_path = '/content/drive/MyDrive/test.csv'

# 데이터 불러오기
train = pd.read_csv(train_path)

# null 값 제거
train = train.dropna()

# 피처와 라벨 분리
features = ['누적전력량', '유효전력평균', '무효전력평균', '주파수', '전류평균', '상전압평균', '선간전압평균', '온도',
            'R상유효전력', 'R상무효전력', 'R상전류', 'R상전압', 'R상선간전압', 'S상유효전력', 'S상무효전력', 'S상전류',
            'S상전압', 'S상선간전압', 'T상유효전력', 'T상무효전력', 'T상전류', 'T상전압', 'T상선간전압']
labels = ['label_역률평균', 'label_전류고조파평균', 'label_전압고조파평균']

X = train[features]
y = train[labels]

# 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 원-핫 인코딩
encoder = OneHotEncoder()
y_encoded = encoder.fit_transform(y).toarray()

# Train-Test 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# 데이터셋 클래스 정의
class CustomDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# 데이터 로더 생성
train_dataset = CustomDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)



# 신경망 모델구성
class MultiOutputModel(nn.Module):
    def __init__(self, input_size, output_size):
        super(MultiOutputModel, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, output_size)
        self.relu = nn.ReLU()


    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)

        return x



# 모델 초기화
input_size = X_train.shape[1]
output_size = y_train.shape[1]
model = MultiOutputModel(input_size, output_size)

# 손실 함수와 옵티마이저
criterion = nn.BCELoss()  # 다중 출력을 갖는 경우 Binary Cross Entropy 손실 함수 선택
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 모델 학습
num_epochs = 50

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss / len(train_loader)}")

# 모델 평가
def evaluate_model(model, test_loader, criterion):
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
    avg_test_loss = test_loss / len(test_loader)
    return avg_test_loss

test_dataset = CustomDataset(X_test, y_test)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
test_loss = evaluate_model(model, test_loader, criterion)
print(f"Test Loss: {test_loss}")

# 학습된 모델 저장
torch.save(model.state_dict(), '/content/drive/MyDrive/multioutput_model0.pth')

# 모델 평가 함수
def evaluate_model(model, test_loader, criterion):
    model.eval()
    test_loss = 0.0
    true_labels = []
    predicted_labels = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            # 예측된 라벨 및 실제 라벨 저장
            predicted_labels.extend((outputs > 0.5).int().tolist())
            true_labels.extend(labels.int().tolist())
    avg_test_loss = test_loss / len(test_loader)
    # 정확도 계산
    accuracy = accuracy_score(true_labels, predicted_labels)
    # 정밀도 계산
    precision = precision_score(true_labels, predicted_labels, average='weighted')
    # 재현율 계산
    recall = recall_score(true_labels, predicted_labels, average='weighted')
    # F1 점수 계산
    f1 = f1_score(true_labels, predicted_labels, average='weighted')
    return avg_test_loss, accuracy, precision, recall, f1

# 평가 함수 호출하여 성능 측정
test_dataset = CustomDataset(X_test, y_test)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
test_loss, accuracy, precision, recall, f1 = evaluate_model(model, test_loader, criterion)

# 성능 출력
print(f"Test Loss: {test_loss}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

```

```python
# 3개 노드 출력 및 얼리스팁

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torch
from torch.utils.data import DataLoader, TensorDataset
import torch.nn as nn
import torch.optim as optim

# 데이터 경로
train_path = '/content/drive/MyDrive/train.csv'

# 데이터 불러오기
train = pd.read_csv(train_path)

# null 값 제거
train = train.dropna()

# 피처와 라벨 분리
features = ['누적전력량', '유효전력평균', '무효전력평균', '주파수', '전류평균', '상전압평균', '선간전압평균', '온도',
            'R상유효전력', 'R상무효전력', 'R상전류', 'R상전압', 'R상선간전압', 'S상유효전력', 'S상무효전력', 'S상전류',
            'S상전압', 'S상선간전압', 'T상유효전력', 'T상무효전력', 'T상전류', 'T상전압', 'T상선간전압']
labels = ['label_역률평균', 'label_전류고조파평균', 'label_전압고조파평균']

X = train[features]
y = train[labels]

# 라벨을 정수 값으로 할당
label_mapping = {'정상': 0, '주의': 1, '경고': 2}
y_encoded = y.applymap(lambda x: label_mapping[x])

# 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PyTorch Tensor로 변환
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_tensor = torch.tensor(y_encoded.values, dtype=torch.long)

# Train-Test 분할
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

# 데이터 로더 생성
train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=2048, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=2048, shuffle=False)

1

# 모델 초기화
model = MyModel()

# 손실 함수와 옵티마이저 설정
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Early Stopping 클래스 정의
class EarlyStopping:
    def __init__(self, patience=5, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_score = None
        self.early_stop = False
        self.counter = 0
        self.best_loss = np.Inf

    def __call__(self, val_loss, model):
        score = -val_loss

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        '''Saves model when validation loss decrease.'''
        self.best_loss = val_loss
        torch.save(model.state_dict(), 'checkpoint.pt')

# Early stopping 객체 생성
early_stopping = EarlyStopping(patience=5, delta=0.01)

# 학습
epochs = 5000
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs1, outputs2, outputs3 = model(inputs)
        loss1 = criterion(outputs1, labels[:, 0])
        loss2 = criterion(outputs2, labels[:, 1])
        loss3 = criterion(outputs3, labels[:, 2])
        loss = loss1 + loss2 + loss3
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs1, outputs2, outputs3 = model(inputs)
            loss1 = criterion(outputs1, labels[:, 0])
            loss2 = criterion(outputs2, labels[:, 1])
            loss3 = criterion(outputs3, labels[:, 2])
            val_loss += loss1.item() + loss2.item() + loss3.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader)}, Val Loss: {val_loss/len(test_loader)}")

    early_stopping(val_loss, model)
    if early_stopping.early_stop:
        print("Early stopping")
        break

print('Finished Training')


# 모델 가중치 저장
model_path = '/content/drive/MyDrive/my_model_final_realy.pth'
torch.save(model.state_dict(), model_path)
print(f"Model saved to {model_path}")



# 모델 가중치 로드 (Early Stopping에서 저장된 최적 가중치)
model.load_state_dict(torch.load('checkpoint.pt'))


# 모델 평가
model.eval()
all_preds1, all_preds2, all_preds3 = [], [], []
all_labels1, all_labels2, all_labels3 = [], [], []

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs1, outputs2, outputs3 = model(inputs)
        _, predicted1 = torch.max(outputs1.data, 1)
        _, predicted2 = torch.max(outputs2.data, 1)
        _, predicted3 = torch.max(outputs3.data, 1)
        all_preds1.extend(predicted1.cpu().numpy())
        all_preds2.extend(predicted2.cpu().numpy())
        all_preds3.extend(predicted3.cpu().numpy())
        all_labels1.extend(labels[:, 0].cpu().numpy())
        all_labels2.extend(labels[:, 1].cpu().numpy())
        all_labels3.extend(labels[:, 2].cpu().numpy())

# 성능 지표 계산
def compute_metrics(all_labels, all_preds):
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
    return accuracy, precision, recall, f1

metrics1 = compute_metrics(all_labels1, all_preds1)
metrics2 = compute_metrics(all_labels2, all_preds2)
metrics3 = compute_metrics(all_labels3, all_preds3)

print('Metrics for label_역률평균 - Accuracy: {:.4f}, Precision: {:.4f}, Recall: {:.4f}, F1 Score: {:.4f}'.format(*metrics1))
print('Metrics for label_전류고조파평균 - Accuracy: {:.4f}, Precision: {:.4f}, Recall: {:.4f}, F1 Score: {:.4f}'.format(*metrics2))
print('Metrics for label_전압고조파평균 - Accuracy: {:.4f}, Precision: {:.4f}, Recall: {:.4f}, F1 Score: {:.4f}'.format(*metrics3))


```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn

# 신경망 모델 정의
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = nn.Linear(23, 256)
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(256, 128)
        self.dropout2 = nn.Dropout(0.5)
        self.fc3 = nn.Linear(128, 64)
        self.fc4_1 = nn.Linear(64, 3)
        self.fc4_2 = nn.Linear(64, 3)
        self.fc4_3 = nn.Linear(64, 3)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout1(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout2(x)
        x = torch.relu(self.fc3(x))
        out1 = self.fc4_1(x)
        out2 = self.fc4_2(x)
        out3 = self.fc4_3(x)
        return out1, out2, out3

# 모델 초기화
model = MyModel()

# 저장된 가중치 로드
model_path = '/content/drive/MyDrive/my_model_final_realy.pth'
model.load_state_dict(torch.load(model_path))
model.eval()

# 훈련 데이터에서 스케일러를 피팅합니다.
train_path = '/content/drive/MyDrive/train.csv'
train = pd.read_csv(train_path)
train = train.dropna()
features = ['누적전력량', '유효전력평균', '무효전력평균', '주파수', '전류평균', '상전압평균', '선간전압평균', '온도',
            'R상유효전력', 'R상무효전력', 'R상전류', 'R상전압', 'R상선간전압', 'S상유효전력', 'S상무효전력', 'S상전류',
            'S상전압', 'S상선간전압', 'T상유효전력', 'T상무효전력', 'T상전류', 'T상전압', 'T상선간전압']
scaler = StandardScaler()
scaler.fit(train[features])

# 새로운 입력 데이터 준비
input_data = {
    '누적전력량': float(input('누적전력량 값 입력: ')),
    '유효전력평균': float(input('유효전력평균 값 입력: ')),
    '무효전력평균': float(input('무효전력평균 값 입력: ')),
    '주파수': float(input('주파수 값 입력: ')),
    '전류평균': float(input('전류평균 값 입력: ')),
    '상전압평균': float(input('상전압평균 값 입력: ')),
    '선간전압평균': float(input('선간전압평균 값 입력: ')),
    '온도': float(input('온도 값 입력: ')),
    'R상유효전력': float(input('R상유효전력 값 입력: ')),
    'R상무효전력': float(input('R상무효전력 값 입력: ')),
    'R상전류': float(input('R상전류 값 입력: ')),
    'R상전압': float(input('R상전압 값 입력: ')),
    'R상선간전압': float(input('R상선간전압 값 입력: ')),
    'S상유효전력': float(input('S상유효전력 값 입력: ')),
    'S상무효전력': float(input('S상무효전력 값 입력: ')),
    'S상전류': float(input('S상전류 값 입력: ')),
    'S상전압': float(input('S상전압 값 입력: ')),
    'S상선간전압': float(input('S상선간전압 값 입력: ')),
    'T상유효전력': float(input('T상유효전력 값 입력: ')),
    'T상무효전력': float(input('T상무효전력 값 입력: ')),
    'T상전류': float(input('T상전류 값 입력: ')),
    'T상전압': float(input('T상전압 값 입력: ')),
    'T상선간전압': float(input('T상선간전압 값 입력: '))
}

new_data = pd.DataFrame([input_data])

# 데이터 표준화
new_data_scaled = scaler.transform(new_data[features])

# PyTorch Tensor로 변환
new_data_tensor = torch.tensor(new_data_scaled, dtype=torch.float32)

# 예측
with torch.no_grad():
    output1, output2, output3 = model(new_data_tensor)
    predicted_prob1 = torch.softmax(output1, dim=1)
    predicted_prob2 = torch.softmax(output2, dim=1)
    predicted_prob3 = torch.softmax(output3, dim=1)

    predicted_class1 = torch.argmax(predicted_prob1, dim=1).item()
    predicted_class2 = torch.argmax(predicted_prob2, dim=1).item()
    predicted_class3 = torch.argmax(predicted_prob3, dim=1).item()

# 클래스 맵핑
class_mapping = {0: '정상', 1: '주의', 2: '경고'}
predicted_label1 = class_mapping[predicted_class1]
predicted_label2 = class_mapping[predicted_class2]
predicted_label3 = class_mapping[predicted_class3]

print(f"Predicted label for label_역률평균: {predicted_label1}")
print(f"Predicted label for label_전류고조파평균: {predicted_label2}")
print(f"Predicted label for label_전압고조파평균: {predicted_label3}")

print(f"Predicted probabilities for label_역률평균: {predicted_prob1.numpy()}")
print(f"Predicted probabilities for label_전류고조파평균: {predicted_prob2.numpy()}")
print(f"Predicted probabilities for label_전압고조파평균: {predicted_prob3.numpy()}")
```
